```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('car.jpeg') 
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(121);plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB));plt.title("Input Image")
plt.subplot(122);plt.imshow(gray_image,cmap='gray');plt.title("Gray Image")
plt.show()
```

```python
blur = cv2.GaussianBlur(gray_image, (5,5), 0)
plt.imshow(blur, cmap='gray'); plt.title("Blurred Image"); plt.show()
```

```python
low = 50
high = 150
edges = cv2.Canny(blur, low, high)
plt.figure(figsize=(10,5))
plt.subplot(1,2,1);plt.imshow(edges, cmap='gray'); plt.title("Canny Edges"); plt.axis('off');
plt.subplot(1,2,2);plt.imshow(gray_image, cmap='gray'); plt.title("Original image");plt.axis('off'); 
plt.show()
```
